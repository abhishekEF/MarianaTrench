Load Census Demographics Data by Tract and remove margin of error data, can be replicable with other state data. Dataset comes from Census [S0701 GEOGRAPHIC MOBILITY BY SELECTED CHARACTERISTICS IN THE UNITED STATES. ](https://data.census.gov/cedsci/table?q=S0701%20%20GEOGRAPHIC%20MOBILITY%20BY%20SELECTED%20CHARACTERISTICS%20IN%20THE%20UNITED%20STATES&g=0400000US27%241400000&tid=ACSST5Y2020.S0701)

```python
df = pd.read_csv("/MinnesotaDemographicstract.csv")
df = df.loc[:, ~df.columns.str.endswith('M')]
df = df[df.columns[~df.columns.str.endswith('M')]]
df = df.loc[:, ~df.columns.str.startswith('S0701_C02')]
df = df[df.columns[~df.columns.str.startswith('S0701_C02')]]
df = df.loc[:, ~df.columns.str.startswith('S0701_C03')]
df = df[df.columns[~df.columns.str.startswith('S0701_C03')]]
df = df.loc[:, ~df.columns.str.startswith('S0701_C04')]
df = df[df.columns[~df.columns.str.startswith('S0701_C04')]]
df = df.loc[:, ~df.columns.str.startswith('S0701_C05')]
df = df[df.columns[~df.columns.str.startswith('S0701_C05')]]
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
```
